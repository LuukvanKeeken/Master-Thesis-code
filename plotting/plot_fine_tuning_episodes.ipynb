```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
percentages = np.linspace(10, 100, 10)
plt.ioff()
fig, ax = plt.subplots()
```

```python
BP_DQRNN_best_episodes_avg, BP_DQRNN_best_episodes_stddev = np.load("../fine_tuning_results/BP_DQRNN_1000eps_best_episodes_correct.npy")
ax.errorbar(percentages, BP_DQRNN_best_episodes_avg, BP_DQRNN_best_episodes_stddev, capsize=5, label = "BP DQRNN")
```

```python
BP_A2C_RNN_best_episodes_avg, BP_A2C_RNN_best_episodes_stddev = np.load("../BP_A2C/fine_tuning_results/FIRST5_BP_A2C_RNN_fine_tuning_result_8_20231016/best_episodes.npy")
ax.errorbar(percentages, BP_A2C_RNN_best_episodes_avg, BP_A2C_RNN_best_episodes_stddev, capsize=5, label = "BP A2C RNN")
```

```python
Standard_A2C_RNN_best_episodes_avg, Standard_A2C_RNN_best_episodes_stddev = np.load("../BP_A2C/fine_tuning_results/Standard_A2C_RNN_fine_tuning_result_3_20231016/best_episodes.npy")
ax.errorbar(percentages, Standard_A2C_RNN_best_episodes_avg, Standard_A2C_RNN_best_episodes_stddev, capsize=5, label = "Standard A2C RNN")
```

```python
ax.plot(percentages, [35.4, 14.6, 60.4, 85.4, 56.3, 62.5, 72.9, 60.4, 47.9, 68.8], label = "r-STDP (estimated from plot)")
```

```python
ax.grid(True)
ax.set_ylim(-30, 250)
ax.set_xlabel("Pole length increase (%)")
ax.set_ylabel("Average fine-tuning episodes")
ax.set_title("Number of fine-tuning episodes needed to reach best evaluation performance")
ax.set_xticks(percentages)
ax.legend()
```
